```python
##UNIVERIATE LINNEAR REGRESSION

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable (2D array)
y = np.array([2, 4, 5, 4, 5])                # Dependent variable

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Print the coefficients
print(f"Slope (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")

# Plotting
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Univariate Linear Regression')
plt.legend()
plt.show()
```

```python
##### GRADIENT DECENT ##########
import numpy as np

# Sample data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Hyperparameters
m = 0  # initial slope
b = 0  # initial intercept
learning_rate = 0.01
epochs = 1000
n = len(X)

# Gradient Descent
for _ in range(epochs):
    y_pred = m * X + b
    error = y - y_pred

    dm = (-2/n) * np.sum(X * error)
    db = (-2/n) * np.sum(error)

    m = m - learning_rate * dm
    b = b - learning_rate * db

print(f"Final slope (m): {m}")
print(f"Final intercept (b): {b}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Sample Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
n = len(X)

# Step 2: Initialize parameters
m = 0  # slope
b = 0  # intercept
learning_rate = 0.01
epochs = 1000

# Step 3: Store loss for visualization
loss_history = []

# Step 4: Gradient Descent Loop
for i in range(epochs):
    y_pred = m * X + b
    error = y - y_pred

    # Calculate gradients
    dm = (-2/n) * np.sum(X * error)
    db = (-2/n) * np.sum(error)

    # Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db

    # Compute and store current loss
    mse = np.mean(error**2)
    loss_history.append(mse)

    # Optional: print every 100 iterations
    if i % 100 == 0:
        print(f"Epoch {i}: MSE = {mse:.4f}, m = {m:.4f}, b = {b:.4f}")

# Step 5: Final result
print("\nFinal model:")
print(f"y = {m:.2f}x + {b:.2f}")

# Step 6: Plot the regression line
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, m * X + b, color='red', label='Predicted Line')
plt.title("Linear Regression (Gradient Descent)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Step 7: Plot the loss over iterations
plt.plot(loss_history)
plt.title("Loss Over Iterations")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Step 1: Load and prepare the dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Choose one feature and one target (let's predict petal length from sepal length)
X = df['sepal length (cm)'].values
y = df['petal length (cm)'].values
n = len(X)

# Reshape X for consistency
X = X.reshape(-1)
y = y.reshape(-1)

# Step 2: Initialize parameters
m = 0  # slope
b = 0  # intercept
learning_rate = 0.01
epochs = 1000
loss_history = []

# Step 3: Define Loss Function (Mean Squared Error)
def compute_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Step 4: Gradient Descent
for epoch in range(epochs):
    y_pred = m * X + b
    error = y - y_pred

    # Gradients
    dm = (-2/n) * np.sum(X * error)
    db = (-2/n) * np.sum(error)

    # Update parameters
    m -= learning_rate * dm
    b -= learning_rate * db

    # Store loss
    loss = compute_loss(y, y_pred)
    loss_history.append(loss)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}, m = {m:.4f}, b = {b:.4f}")

# Step 5: Final Model
print("\nFinal Model:")
print(f"y = {m:.2f}x + {b:.2f}")

# Step 6: Visualization

# Plot 1: Regression Line
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, m * X + b, color='red', label='Regression Line')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Univariate Linear Regression on Iris Dataset')
plt.legend()
plt.grid(True)

# Plot 2: Loss Curve
plt.subplot(1, 2, 2)
plt.plot(loss_history)
plt.title('Loss Over Iterations')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.grid(True)

plt.tight_layout()
plt.show()
```
