```python
import pandas as pd
import re

file_path = r"C:\Users\RF\Desktop\test.csv"
data = pd.read_csv(file_path)

data = data[['description_x']].copy()

# Define a function to clean the text
def clean_description(text):
    if pd.isnull(text):
        return text
    # Remove ISIN (e.g., US00206R1023)
    text = re.sub(r'\b[A-Z]{2}[A-Z0-9]{9}\b', '', text)
    # Remove SEDOL (e.g., #2831811)
    text = re.sub(r'#\d{7}', '', text)
    # Remove other unwanted symbols or extra spaces
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    text = re.sub(r'\s+', ' ', text)  # remove extra whitespace
    return text.strip()

# Create a new column for cleaned descriptions
data['cleaned_description'] = data['description_x'].apply(clean_description)

# Show the result
data.tail()
```
